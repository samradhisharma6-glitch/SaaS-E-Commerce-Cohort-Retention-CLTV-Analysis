```python
missing = (df.isnull().sum()/len(df))*100

missing.sort_values(ascending=False)
```

```python
#Remove Invalid Customer Records
```

```python
df = df.dropna(subset=['Customer ID'])
```

```python
df.shape
```

```python
#Remove Cancelled Orders
```

```python
df[df['Invoice'].astype(str).str.startswith('C')].head()
```

```python
df = df[~df['Invoice'].astype(str).str.startswith('C')]
```

```python
df.shape
```

```python
#Remove Negative Quantity
df[df['Quantity'] <= 0]
```

```python
df = df[df['Quantity'] > 0]
```

```python
#Remove Negative Price
df = df[df['Price'] > 0]
```

```python
df[['Quantity','Price']].describe()
```

```python
#Remove Duplicates
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
df.duplicated().sum()
```

```python
#Convert Date Format
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df.info()
```

```python
timestamp = np.datetime64('2026-07-07', 'ns')
```

```python
#Create Revenue Column
df['Revenue'] = df['Quantity'] * df['Price']
df[['Quantity','Price','Revenue']].head()
```

```python
#Create Purchase Month
df['InvoiceMonth'] = df['InvoiceDate'].dt.to_period('M')
df[['InvoiceDate','InvoiceMonth']].head()
```

```python
#Create Cohort Month
cohort = df.groupby('Customer ID')['InvoiceMonth'].min()
df['CohortMonth'] = df['Customer ID'].map(cohort)
```

```python
df[['Customer ID','InvoiceMonth','CohortMonth']].head()
```

```python
df.to_csv(
    "online_retail_cleaned.csv",
    index=False
)
print ("successfully file saved")
```

```python
# Ensure raw_df and df are defined before running this
if 'raw_df' in locals() and 'df' in locals():
    raw_rows = len(raw_df)
    clean_rows = len(df)

    removed = raw_rows - clean_rows
    retention = (clean_rows / raw_rows) * 100 if raw_rows > 0 else 0

    print(f"Removed Rows: {removed}")
    print(f"Retention %: {round(retention, 2)}%")
else:
    print("Error: Ensure both 'raw_df' and 'df' are defined.")
```
