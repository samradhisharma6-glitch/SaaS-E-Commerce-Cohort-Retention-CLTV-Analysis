```python
#Create Purchase Month
# 1. Force Pandas to create a completely independent copy of the data
df = df.copy()

# 2. Ensure date format is correct
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# 3. Create the InvoiceMonth column
df['InvoiceMonth'] = df['InvoiceDate'].dt.to_period('M')

# 4. Create the CohortMonth column
cohort = df.groupby('Customer ID')['InvoiceMonth'].min()
df['CohortMonth'] = df['Customer ID'].map(cohort)
```
